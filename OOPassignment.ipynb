```python
class Book():
    def __init__(self,title, author, isbn):
        self.title = title
        self.author = author
        self.isbn   = isbn
        self.is_checked_out = False
    def display_info(self):
        print(self.title)
        print(self.author)
        print(self.isbn)
        print("--fauji brand library--")
    def Checked_out(self):
        if self.is_checked_out:
          print(f"[{self.title}]This book is checked out")
        else:
            self.is_checked_out =True
            print(f"[{self.title}]This book has been checked out(taken)")
    def check_in(self):
        if self.is_checked_out:
            self.Checked_out = False
            print(f"[{self.title}] has been checked in(mojood)")
        else:
             print(f"[{self.title}] is available ")

#object creation

book1 = Book("Harry Potter", "J.K.Rowling", "123")
book2 = Book("Lion & Wolf", "Rex Holloway", "456")
book3 = Book("Farmer & 3 Sons", "Chantal de Marolles", "789")
book4 = Book("Wood Cutter", "Thomas Bernhard", "101112")
book5 = Book("The Jungle Book", "Rudyard Kipling", "131415")
book6 = Book("The Haunted House", "Charles Dickens", "161718")
#testing of class Book
book2.display_info()
book6.display_info()
book1.Checked_out()
book1.Checked_out()
book1.check_in()
book3.check_in()
```

```python
class Library:
    def __init__(self, name):
        self.name = name
        self.books = []
    def add_book(self, book_object):
        self.books.append(book_object)
        print(f"[{book_object.title}] has been added to {self.name} library.")

    def find_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        print(f"Book with title '{title}' not found.")
        return None
    def check_out_book(self, title):
        book = self.find_book(title)
        if book:
            book.check_out()

    def check_in_book(self, title):
        book = self.find_book(title)
        if book:
            book.check_in()

    def list_all_books(self):
        print(f"Available books in {self.name}:")
        if self.books:
            for book in self.books:
                book.display_info()
        else:
            print("No books in library.")
             # Creating book objects
book1 = Book("Harry Potter", "J.K. Rowling", "123")
book2 = Book("The Hobbit", "J.R.R. Tolkien", "456")
book3 = Book("Charlie & Chocolate Factory", "Roald Dahl", "789")

# Creating library and adding books
library = Library("Wonderful Library")
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Listing all books
library.list_all_books()
```
